```python
import json

# Load the original dataset
file_path = './matched_fd.json'

with open(file_path, 'r') as file:
    data = json.load(file)

```

```python
from random import randint, choice

def generate_adversarial_text(text):
    # Simple text modifications for demonstration
    synonyms = {
        'adenocarcinoma': 'glandular cancer',
        'lymphovascular invasion': 'vascular infiltration by cancer',
        'perineural invasion': 'nerve invasion',
        'malignancy': 'malignant tumor',
        'biopsy': 'tissue sampling',
        'metastasis': 'spread of cancer',
        'negative for malignancy': 'no cancer detected',
        'positive for metastasis': 'cancer spread confirmed',
        'benign': 'non-cancerous',
        'fibrofatty': 'fibroadipose tissue',
        'lymphocytic infiltrate': 'lymphocyte infiltration',
        'necrosis': 'tissue death',
        'hyperplasia': 'excessive cell growth',
        'dysplasia': 'abnormal cell growth',
        'carcinoma': 'malignant epithelial tumor',
        'sarcoma': 'malignant tumor of connective tissue',
        'melanoma': 'skin cancer',
        'leukemia': 'blood cancer',
        'margins': 'surgical borders',
        'histologic grade': 'tumor grade',
        'staging': 'classification of cancer extent',
        'in situ': 'localized tumor',
        'invasive': 'infiltrating cancer',
        'hemorrhage': 'bleeding',
        'edema': 'swelling from fluid accumulation',
        'fibrosis': 'scarring process',
        'angiogenesis': 'formation of new blood vessels',
        'cyst': 'fluid-filled sac',
        'nodule': 'small lump',
        'polyp': 'growth protruding from a mucous membrane',
        'granuloma': 'small area of inflammation',
        'oncology': 'study of tumors',
        'radiology': 'medical imaging study',
        'chemotherapy': 'drug treatment for cancer',
        'radiation therapy': 'cancer treatment using radiation',
        'immunotherapy': 'treatment to boost immunity',
        'surgical resection': 'surgical removal of tissue',
        'diagnostic': 'identifying disease',
        'prognosis': 'disease outcome prediction',
        'palliative care': 'relieving symptoms without curing',
        'recurrence': 'return of disease',
        'remission': 'reduction of disease symptoms',
        'biological therapy': 'treatment with biological agents',
        'targeted therapy': 'specific therapeutic approach',
        'laparotomy': 'surgical opening of the abdomen',
        'endoscopy': 'internal examination using an endoscope',
        'biomarker': 'biological indicator of disease state',
        'cytotoxic': 'toxic to cells',
        'neoadjuvant therapy': 'treatment before primary therapy',
        'adjuvant therapy': 'additional treatment'
    }
    
    # Numeric alteration function
    def alter_number(match):
        num = float(match.group())
        # Randomly increase or decrease the number by up to 10%
        deviation = randint(-10, 10) / 100
        new_num = num + (num * deviation)
        # Maintain the original number's precision
        if "." in match.group():
            return "{:.1f}".format(new_num)
        else:
            return str(int(new_num))
    
    # Apply synonym replacements
    for original, synonym in synonyms.items():
        if original in text:
            text = text.replace(original, choice([original, synonym]))
    
    # Apply numeric alterations
    import re
    text = re.sub(r'\b\d+\.\d+\b', alter_number, text)  # Alter decimal numbers
    text = re.sub(r'\b\d+\b', alter_number, text)  # Alter integer numbers

    return text

# Generate adversarial examples for a subset of the dataset
adversarial_dataset = []
for record in data[501:2503]:  # Limiting to first 10 records for demonstration
    modified_record = {}
    for key, value in record.items():
        if isinstance(value, str):  # Apply changes only to textual data
            modified_value = generate_adversarial_text(value)
            modified_record[key] = modified_value
        else:
            modified_record[key] = value
    adversarial_dataset.append(modified_record)

# # Display the first modified record for review
# adversarial_dataset[0]
```

```python
with open('adversarial_dataset.json', 'w') as file:
    json.dump(adversarial_dataset, file, indent=4)
```
